```python
book1=book.drop_duplicates()
book1
```

```python
book1.isnull().sum() #null values
```

```python
#number of unique users in the dataset
len(book1.user_id.unique())
```

```python
len(book1.Book_Title.unique())
```

```python
book2 = book1.pivot_table(index='user_id',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
```

```python
book2
```

```python
book2.index = book1.user_id.unique()
```

```python
book2.head(30)
```

```python
#Impute those NaNs with 0 values
book2.fillna(0, inplace=True)
```

```python
book2.head(30)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book2.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book1.user_id.unique()
user_sim_df.columns = book1.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
